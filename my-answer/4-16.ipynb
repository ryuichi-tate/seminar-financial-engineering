# 問 4-16 モーゲージの分割

## (a)

\begin{align}
V = \frac{n}{1+r}(B-rM)
\end{align}

## (b)

\begin{align}
V = \frac{r}{1+r}\frac{nM}{(1+r)^n-1}
\end{align}

## (c)

\begin{align}
W = M-V
\end{align}

## (e)

元本のデュレーション
\begin{align}
D_V &= -\frac{1}{V}\frac{dV}{dr}\\
&= -\frac{1}{r(1+r)}+\frac{n(1+r)^{n-1}}{(1+r)^n-1}
\end{align}

利息のデュレーション
\begin{align}
D_W &= -\frac{1}{W}\frac{dW}{dr}\\
&= -(n+1)\frac{(1+r)^n-1}{(1+r)^{n+1}-(n+1)r-1} + \frac{1}{1+r} + \frac{n(1+r)^{n-1}}{(1+r)^n-1}
\end{align}

デュレーションの差
\begin{align}
D_V-D_W &= -\frac{1}{r} + (n+1)\frac{(1+r)^n-1}{(1+r)^{n+1}-(n+1)r-1} \\
&= f(r;n)
\end{align}

$y=f(r;n)$のグラフを描く

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def Dv(r,n):
    return -1/(r*(1+r))+(n*(1+r)**(n-1))/((1+r)**n-1)
```

```python
def Dw(r,n):
    return -(n+1)*((1+r)**n-1)/((1+r)**(n+1)-(n+1)*r-1) + 1/(1+r) + n*(1+r)**(n-1)/((1+r)**n-1)
```

```python
def f(r,n):
    return -1/r + (n+1)*((1+r)**n-1)/((1+r)**(n+1)-(n+1)*r-1)
```

```python
r=np.arange(0,1,0.01)
for n in range(1,21,5):
    plt.plot(r, f(r,n=n),label='$n={0}$'.format(n))
plt.plot(r,np.ones_like(r)*0, label='zero line')
plt.legend()
plt.xlabel('$r$')
plt.ylabel('$f(r;n)$\n$=D_V-D_W$')
plt.show()
```

```python
r=np.arange(0,1,0.01)
for n in range(1,21,5):
    plt.plot(r, Dv(r,n=n),label='$n={0}$'.format(n))
plt.plot(r,np.ones_like(r)*0, label='zero line')
plt.legend()
plt.xlabel('$r$')
plt.ylabel('$D_V$')
plt.show()
```

```python
r=np.arange(0,1,0.01)
for n in range(1,21,5):
    plt.plot(r, Dw(r,n=n),label='$n={0}$'.format(n))
plt.plot(r,np.ones_like(r)*0, label='zero line')
plt.legend()
plt.xlabel('$r$')
plt.ylabel('$D_W$')
plt.show()
```
